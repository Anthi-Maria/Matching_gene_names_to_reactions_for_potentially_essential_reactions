```python
i= -1

J=list()
C=list()
J.append(a.genes[0].name)
C.append(a.genes[0].reactions)

dictionary = dict(zip(J, C))

while i <1344:

    J= list()
    C= list()

    J.append(a.genes[i].name)
    C.append(a.genes[i].reactions)

    dictionary2 = dict(zip(J, C))
    dictionary.update(dictionary2)

    i=i+1

dictionary
```

```python
import csv

with open('dictionary.csv', 'w') as f:
    writer = csv.writer(f)
    for k, v in dictionary.items():
       writer.writerow([k, v])
```
